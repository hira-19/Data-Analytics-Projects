### Creating a separate, smaller dataframe

```python
df_new = df.loc[:100, ['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department']]
```

```python
df_new
```

### Adding a column


```python
#will automatically create a column if it doesn't exist and add it at the end

df_new["new_col"] = df.loc[1:100, 'JobRole']

df_new
```

```python
#adding a column at a specific location
def_new = df_new.insert(loc= 1, column = "new_col02", value = df["Department"])
df_new
```

### Dropping a column

```python
#to drop columns, you need to set axis = 1. by default it is set to 0, meaning rows
#this is not an inplace change (i.e. a copy is returned instead of the original df_new changing) 


df_new = df_new.drop("new_col", axis = 1)
df_new
```

```python
# if you change the setting inplace = True then original df_new will be changed inplace

# df_new.drop("new_col3", axis = 1, inplace= True) #axis 1 for col and 0 for rows
df_new
```

### Nulls

```python
#checking nulls in whole df_new


df_new.isnull()
```

```python
#here too you can change the inplace setting to return a copy or make the change to original df_new
#with inplace = False right now, we are getting a copy but no changes are made to the original

df_new.fillna(0, inplace = False)
```

```python
#checking total nulls

df_new["Age"].isnull().sum()
```

### Using .apply()

```python
#applying a string method without .apply(). this is returning a copy

df["BusinessTravel"].str.upper()
```

```python
#applying a string method with .apply(). this is returning a copy

df["BusinessTravel"].apply(str.upper)
```

```python
#this lambda function is taking each row's string as input and returning the first letter. 

df["BusinessTravel"].apply(lambda x: x[0]) 
```

```python
#concatenating with the plus operator. this is just a copy being returned. no inplace changes

df["BusinessTravel"].apply(lambda x: x[0]) + df["Department"].apply(lambda x: x[0]) 
```

### using .agg to apply functions

```python
df.groupby('Department')[["Age", "DistanceFromHome"]].agg(np.mean)
```

```python
df.groupby('Department')[["Age", "DistanceFromHome"]].agg('count')
```

```python
#passing a dictionary to .agg to apply specific aggregate functions to specific columns

df.groupby("JobRole").agg({"Age": [np.mean, np.min], "HourlyRate": [np.max , 'count']})
```

```python
#with multiple groupings

df.groupby(["JobRole", "Age"]).agg({"Age": [np.mean, np.min], "HourlyRate": [np.max , 'count']})
```
